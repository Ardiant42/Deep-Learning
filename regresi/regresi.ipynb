# Feature Selection

- We use correlation matrix for feature selection

```python
sns.heatmap(df.corr(),annot=True, cmap="Greens")
```

- From correlation heatmap we can see that 'house price' is highly correlated with features 'x2 house age' 'X3 distance to the nearest MRT station' 'X4 number of convenience stores'	'X5 latitude'	'X6 longitude'.
- So we can take these features to predict  'house price of unit area'.

```python
# Independent features
x = df.drop(['No','X1 transaction date'], axis=1)

# Dependent featurs
y = df['Y house price of unit area']
```

```python
# Creating training and testing datasets

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

# Training the model using linear regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
# Accuracy score of the model

model.score(x_test, y_test)
```
